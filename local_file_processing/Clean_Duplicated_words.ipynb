```python

def SortList(input_file_path):
    words = []
    with open(input_file_path, 'r', encoding='utf-8') as file:
        content = file.read()
        
    lines = content.splitlines()
    for line in lines:
        words.append(line.split(' '))
    

    word_list = [(item[0].split(':')[0], int(item[0].split(':')[1])) for item in words]
    # sorted_list = sorted(word_list, key=lambda x: (len(x[0]),x[0],x[1]))
    return word_list

sorted_list_MHD = SortList("../wordcount/aggregate_wordcount.txt")
Moderner_List = SortList("../wordcount/aggregate_Cleaned.txt")

List_With_Index = []
for (index, item) in enumerate(sorted_list_MHD):
    key,value = item
    new_item = (index,key,value)
    List_With_Index.append(new_item)

print(List_With_Index[100])
print(Moderner_List)

```

```python
def find_containing_words(word_list):
    containing_words = set()
    
    words_only = [word for _,word, _ in word_list if len(word) > 3]  
    
    for _, word, _ in word_list:
        if len(word) > 4 and any(w in word and w != word for w in words_only if w != word):
            containing_words.add(word)
    
    return list(containing_words)

def not_contain_other_words(word_list):
    containing_words = find_containing_words(word_list)
    short_words = [word for _,word, _ in word_list if (len(word) <= 3 and len(word) >= 2) ]
    result = [word for _, word, _ in word_list if word not in containing_words]
    final_result = [(number, word, freq) for number, word, freq in word_list if word in short_words or word not in containing_words]
    return final_result

def clean_index(word_list):
    for (item) in enumerate(sorted_list_MHD):
        index,key,value = item
        new_item = (key,value)
    List_With_Index.append(new_item)

```

```python
def save_results_to_file(results, filename):
    with open(filename, 'w') as file:
        for number, word, freq in results:
            file.write(f"{number},{word}: {freq}\n")

def save_results_to_file2(results, filename):
    with open(filename, 'w') as file:
        for  word, freq in results:
            file.write(f"{word}: {freq}\n")
```
