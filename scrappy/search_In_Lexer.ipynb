```python
import requests 
from bs4 import BeautifulSoup
```

```python
original_words = []
new_words = []
def cleaningnumber(input_file_path,list):
    with open(input_file_path, 'r', encoding='utf-8') as file:
        content = file.read()
        
    lines = content.splitlines()
    for line in lines:
        list.append(line.split(':')[0])
    
    return list

original_input_file_path = '../wordcount/MHD_NoDuplicate.txt'
cleaningnumber(original_input_file_path,original_words)
```

```python

def LexerSearch(raw):
    # URL for the API request
    url = f"https://api.woerterbuchnetz.de/dictionaries/Lexer/lemmata/lemma/{raw}*/100/json"

# Query parameters
    params = {
        'term': 'dac',
        '_type': 'query', 
        'q': 'dac'
    }

# Headers to include in the request
    headers = {
        'Sec-Ch-Ua': '"Not;A=Brand";v="24", "Chromium";v="128"',
        'Accept': 'application/json, text/javascript, */*; q=0.01',
        'Accept-Language': 'de-DE,de;q=0.9',
        'Sec-Ch-Ua-Mobile': '?0',
        'User-Agent': 'Mozilla/5.0 (Windows NT 10.0; Win64; x64) AppleWebKit/537.36 (KHTML, like Gecko) Chrome/128.0.6613.120 Safari/537.36',
        'Sec-Ch-Ua-Platform': '"macOS"',
        'Origin': 'https://woerterbuchnetz.de',
        'Sec-Fetch-Site': 'same-site',
        'Sec-Fetch-Mode': 'cors',
        'Sec-Fetch-Dest': 'empty',
        'Referer': 'https://woerterbuchnetz.de/',
        'Accept-Encoding': 'gzip, deflate, br',
        'Priority': 'u=1, i',
        'Connection': 'keep-alive'
    }

# Send the GET request
    response = requests.get(url, headers=headers, params=params)

# Check if the request was successful
    if response.status_code == 200:
    # Parse the JSON response
        data = response.json()
        # print(data)
    else:
        print(f"Request failed with status code: {response.status_code}")


#now deal with responce part check : 
    filtered_data = [item for item in data if item.get('label') == raw]
    if not filtered_data and not data:
        filtered_data = [{'value': 'A0000', 'label': 'NONE'}]
    elif not filtered_data and data:
        filtered_data = [data[0]]  # Keep the first item

    # print(filtered_data)
    
    if filtered_data:
        result_value = filtered_data[0].get('value', {'value':'A0000'})  # Return the value or 'A0000' if not present
    else:
        result_value = 'A00000'  # Return 'A0000' if the list is empty
    # index_list = []
    # index_list.append(result_value)
    print(result_value)
    return result_value

```

```python
with open('./indexlistxaf.txt', 'w') as file:
    for word in original_words:
        temp = LexerSearch(word)
        file.write(str(temp))
        file.write('\n') 
    print('done')
```
